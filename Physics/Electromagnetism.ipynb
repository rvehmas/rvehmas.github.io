# ⚡️ Electromagnetism

## Gravity’s Spicy Cousin 🌶️🪐

Imagine two small objects floating in space. Gravity pulls them together with a force that weakens as they move apart — specifically, it fades with the *square* of the distance. Electric forces behave in a similar way: they also follow an **inverse-square law**, becoming weaker as the distance grows. But here comes the twist. While gravity only attracts, electricity comes in two “flavors” of matter. We call them **positive** and **negative** electric charge.  

Two positive charges push each other away, and two negatives do the same. But mix a positive and a negative, and they pull together with remarkable strength.  In fact, the electric force is **immensely stronger** than gravity. Between two electrons, the electric repulsion is about \(10^{39}\) times stronger than their mutual gravitational attraction — an almost absurd difference!

So why don’t we notice this overwhelming electric force in everyday life? Because the matter around us is built from atoms and molecules that balance their charges perfectly. Each atom carries equal amounts of positive and negative charge, and the effects cancel almost completely. The world we see is, by and large, **electrically neutral**, so gravity gets to dominate the large-scale stage.

---

## Where the Charges Hide 🧱🔋

Inside every atom sits a tiny nucleus packed with **protons** (positively charged) and **neutrons** (neutral). Surrounding it is a light, fast cloud of **electrons**, each carrying a negative charge. Normally, the number of electrons equals the number of protons, so the total charge of the atom is zero.  

But it’s surprisingly easy to upset this balance. Rub a balloon on your sweater, and a few electrons hop from one surface to the other. The balloon now has a small excess of charge — and you can feel the consequences when it clings to the wall or makes your hair stand on end. ⚡💇‍♀️

That’s the electric force in action: tiny imbalances of charge, magnified by nature’s powerful law of attraction and repulsion.

---

## The Space Around a Charge 🌊✨

A single charge changes the “mood” of the space around it. We call this invisible influence an **electric field**. If you could place a tiny positive “test charge” nearby, it would feel a push or pull depending on the sign of the source. We often draw lines to represent this field: they flow outward from positive charges and inward toward negative ones. Where the lines are close together, the field is strong; where they spread apart, it weakens.

When several charges are nearby, their fields simply add up, creating intricate patterns that twist and bend. The result looks like a graceful dance of invisible threads filling space. 🩰

---

## Conductors, Insulators, and the Freedom of Electrons 🧰🔧

In some materials, such as metals, certain electrons are not tightly bound to individual atoms. They can move freely from place to place. These materials are called **conductors**, because charges can flow through them easily, redistributing themselves until balance is reached.  Other materials, like rubber or glass, hold their electrons tightly in place. These are **insulators**, and charge tends to stay where it was put.

Between the two extremes lie **semiconductors**, where the freedom of electrons can be finely tuned. Our modern electronic world — computers, solar cells, and LEDs — is built on controlling this delicate balance between conduction and insulation.

---

## Neutral but Not Indifferent 🧲➡️⬅️

Even an electrically neutral object can *respond* to nearby charges. When a charged balloon approaches a wall, the electrons and protons in the wall shift slightly in opposite directions. The wall doesn’t gain or lose charge overall — it merely rearranges them internally.  This subtle effect, called **polarization**, creates a faint positive region near one side and a negative region near the other. The result is a gentle attraction between the charged balloon and the neutral wall.

Neutral, therefore, doesn’t mean uninvolved. Matter is full of mobile charges waiting to adjust their positions in response to what’s nearby.

---

## Voltage: A Landscape for Charges 🗺️🌄

Imagine hiking over hills and valleys. A ball placed high on a hill will roll downhill because of gravity. Electric charges behave in much the same way, except their “height” is called **electric potential**, or **voltage**. 

A **battery** is like a tiny mountain pump: it keeps one end “high” and the other “low” in electric potential. Connect the two ends with a conductor, and charges begin to flow — that is what we call **electric current**. Some devices, like **capacitors**, store energy by keeping charges separated across a thin insulating layer, rather like water stored behind a dam.

---

## Magnetism: The Dance of Moving Charges 🚴‍♀️🧲💫

Magnetism is not a separate phenomenon from electricity — it’s really a part of it. Whenever charges move, they generate magnetic effects. A flowing current in a wire produces a **magnetic field** that curls around it, while electrons spinning inside a magnet create permanent magnetic poles.

In fact, the **total electromagnetic force** on a moving charge depends on three things:  
👉 its **position** (where it is in the field),  
👉 its **velocity** (how it’s moving), and  
👉 its **charge** (how much and what kind).  

A charge that is standing still feels only the electric part, but once it starts to move, magnetism appears — as if motion itself awakens a hidden companion force. ✨

Electricity and magnetism are inseparable partners. Changing an electric field produces a magnetic one, and changing a magnetic field creates an electric one. Together they form the elegant unity of **electromagnetism** — and the most astonishing offspring of this union is **light itself**.  Light is nothing more than a self-sustaining wave of electric and magnetic fields, racing through empty space at the speed of light. 🌈⚡

---

## What Are Electric and Magnetic Fields, Really? 🌍📈

Think of electric and magnetic fields as the *forces that would be felt by a charge* if you placed it at a certain spot, at a certain time.  At every point in space, and at every instant, there are two vector quantities — **E** (the electric field) and **B** (the magnetic field).  They tell us the direction and strength of the push that a charge would experience there.

A “field” in physics simply means any quantity that takes on different values at different points in space — like temperature, pressure, or wind velocity. But here, the field describes something far more fundamental: how the fabric of space itself responds to charge and motion.

The relationships between the values of **E** and **B** at one point and at nearby points are surprisingly simple.  With only a few elegant connections — written as differential equations — we can describe the fields completely. 🌐✏️  

Those relationships are known as **Maxwell’s equations**, and they are among the most beautiful summaries of nature ever discovered. In just four concise statements, they describe how charges and currents create electric and magnetic fields, and how those fields in turn create light, radio waves, and all electromagnetic phenomena we know.

---

## Why Forces Fade with Distance 🌀📏

Suppose a charge sends out its influence equally in every direction, like the ripples from a stone dropped in water. As the ripples expand, the same total “influence” must spread over a larger and larger sphere.  The area of that sphere grows as the square of its radius, so each patch of space receives a smaller share of the effect.  That’s why both gravity and electric forces follow the same **inverse-square** pattern — a direct consequence of simple three-dimensional geometry.

---



```python
import numpy as np
import matplotlib.pyplot as plt

# --- field of a single point charge at the origin ---
def e_field_point(x, y, q=1.0, x0=0.0, y0=0.0, k=1.0, eps=1e-6):
    dx = x - x0
    dy = y - y0
    r2 = dx*dx + dy*dy + eps
    r3 = r2*np.sqrt(r2)
    Ex = k*q*dx/r3
    Ey = k*q*dy/r3
    return Ex, Ey

# grid
xs = np.linspace(-3, 3, 25)
ys = np.linspace(-3, 3, 25)
X, Y = np.meshgrid(xs, ys)

Ex, Ey = e_field_point(X, Y, q=1.0)

plt.figure(figsize=(6,6), constrained_layout=True)
plt.quiver(X, Y, Ex, Ey, pivot='mid', angles='xy', scale=25)
plt.axhline(0, color='0.8'); plt.axvline(0, color='0.8')
plt.title("Electric field vectors around a + charge")
plt.xlabel("x"); plt.ylabel("y"); plt.gca().set_aspect('equal', 'box')
plt.show()

```

```python
def build_electric_field_demo():
    """
    Voilá-ready interactive visualization of electric fields.

    What it shows (aligned with your narrative):
      • Vectors: little arrows that give the field E's direction and strength at many points.
      • Field lines (streamlines): curves everywhere tangent to E; line thickness/color hint at strength.
        (This mirrors the “lines closer together when stronger” convention.)

    Controls:
      - Presets for 1–4 point charges (signs/positions).
      - Toggle vectors / field lines / magnitude background.
      - Density and grid resolution.
      - Clip radius to avoid singularities near charges.
      - Optional normalization to emphasize direction over magnitude.

    Returns:
      ipywidgets.VBox suitable for Voilá dashboards.
    """
    import numpy as np
    import matplotlib.pyplot as plt
    from ipywidgets import (
        VBox, HBox, Dropdown, FloatSlider, IntSlider, Checkbox, Layout, Output, Label
    )
    from IPython.display import clear_output, display

    # -----------------------
    # Physics helpers
    # -----------------------
    def e_field_multi(x, y, charges, k=1.0, eps=1e-6):
        Ex = np.zeros_like(x, dtype=float)
        Ey = np.zeros_like(y, dtype=float)
        for (x0, y0, q) in charges:
            dx = x - x0
            dy = y - y0
            r2 = dx*dx + dy*dy + eps
            r3 = r2*np.sqrt(r2)
            Ex += k*q*dx/r3
            Ey += k*q*dy/r3
        return Ex, Ey

    # -----------------------
    # UI widgets
    # -----------------------
    preset = Dropdown(
        options=[
            ("Single + at center", "single_plus"),
            ("Single − at center", "single_minus"),
            ("Dipole (+ at −1,0; − at +1,0)", "dipole"),
            ("Two like charges (+ +)", "two_plus"),
            ("Quadrupole (+ − + −)", "quad"),
        ],
        value="dipole",
        description="Preset:",
        layout=Layout(width="310px")
    )
    extent = FloatSlider(value=3.0, min=1.5, max=6.0, step=0.5, description="Extent:", readout_format=".1f")
    gridN  = IntSlider(value=25, min=15, max=45, step=2, description="Grid:")
    density = FloatSlider(value=1.4, min=0.5, max=2.5, step=0.1, description="Lines dens.:")
    show_vec = Checkbox(value=True, description="Show vectors (arrows)")
    show_lines = Checkbox(value=True, description="Show field lines")
    show_mag = Checkbox(value=False, description="Show |E| background")
    normalize = Checkbox(value=False, description="Normalize arrows (direction only)")
    clip_r = FloatSlider(value=0.10, min=0.0, max=0.5, step=0.01, description="Clip r<")
    out = Output(layout=Layout(border="1px solid #ddd"))

    # -----------------------
    # Charge presets
    # -----------------------
    def get_charges(name):
        if name == "single_plus":
            return [(0.0, 0.0, +1.0)]
        if name == "single_minus":
            return [(0.0, 0.0, -1.0)]
        if name == "dipole":
            return [(-1.0, 0.0, +1.0), (1.0, 0.0, -1.0)]
        if name == "two_plus":
            return [(-1.0, 0.0, +1.0), (1.0, 0.0, +1.0)]
        if name == "quad":
            return [(-1.0, 0.0, +1.0), (0.0, -1.0, -1.0), (1.0, 0.0, +1.0), (0.0, 1.0, -1.0)]
        return [ (0.0, 0.0, +1.0) ]

    # -----------------------
    # Drawing
    # -----------------------
    def recompute_and_draw(*_):
        with out:
            clear_output(wait=True)
            L = extent.value
            xs = np.linspace(-L, L, gridN.value)
            ys = np.linspace(-L, L, gridN.value)
            X, Y = np.meshgrid(xs, ys)
            charges = get_charges(preset.value)

            # Avoid singularities by clipping very near each charge
            # (We simply blank the field inside r < clip_r for plotting stability.)
            mask = np.ones_like(X, dtype=bool)
            for (x0, y0, q) in charges:
                r = np.hypot(X - x0, Y - y0)
                if clip_r.value > 0:
                    mask &= (r >= clip_r.value)

            Ex, Ey = e_field_multi(X, Y, charges)
            E = np.hypot(Ex, Ey)
            Ex_plot, Ey_plot = np.copy(Ex), np.copy(Ey)
            Ex_plot[~mask] = 0.0; Ey_plot[~mask] = 0.0

            fig = plt.figure(figsize=(7.5, 6.5), constrained_layout=True)
            ax = fig.add_subplot(1,1,1)
            ax.set_aspect('equal', 'box')
            ax.set_xlim(-L, L); ax.set_ylim(-L, L)
            ax.set_xlabel("x"); ax.set_ylabel("y")
            ax.grid(False)

            # Optional |E| background
            if show_mag.value:
                im = ax.imshow(
                    np.flipud(E),
                    extent=[-L, L, -L, L],
                    interpolation='bilinear',
                    cmap='magma',
                    alpha=0.70
                )
                cbar = fig.colorbar(im, ax=ax, fraction=0.046, pad=0.04)
                cbar.set_label("|E| (relative)")

            # Field lines (streamplot) with linewidth tied to |E|
            if show_lines.value:
                # a slightly denser underlying grid for smoother lines
                xs_f = np.linspace(-L, L, max(180, gridN.value*4))
                ys_f = np.linspace(-L, L, max(180, gridN.value*4))
                Xf, Yf = np.meshgrid(xs_f, ys_f)
                Exf, Eyf = e_field_multi(Xf, Yf, charges)
                Ef = np.hypot(Exf, Eyf)

                # line width ∝ |E| (softly normalized)
                lw = 1.5 * (Ef / (Ef.max() + 1e-12)) + 0.25
                sp = ax.streamplot(
                    Xf, Yf, Exf, Eyf,
                    density=density.value,
                    color=Ef,
                    linewidth=lw,
                    cmap='viridis',
                    arrowsize=1.2
                )
                if not show_mag.value:
                    cb = fig.colorbar(sp.lines, ax=ax, fraction=0.046, pad=0.04)
                    cb.set_label("|E| (relative)")

            # Vectors (quiver)
            if show_vec.value:
                U, V = Ex_plot, Ey_plot
                if normalize.value:
                    UU = U / (E + 1e-12)
                    VV = V / (E + 1e-12)
                    UU[~mask] = 0.0; VV[~mask] = 0.0
                    ax.quiver(X, Y, UU, VV, pivot='mid', angles='xy', scale=30, alpha=0.9)
                else:
                    # autoscale based on grid extent
                    scale = 30 * (25 / gridN.value)
                    ax.quiver(X, Y, U, V, pivot='mid', angles='xy', scale=scale, alpha=0.9, color='black')

            # Draw charges
            for (x0, y0, q) in charges:
                ax.scatter(x0, y0, s=230, c=('crimson' if q>0 else 'royalblue'),
                           edgecolor='white', linewidths=1.6, zorder=4)
                ax.text(x0, y0+0.22, '+' if q>0 else '–', ha='center', va='bottom',
                        fontsize=14, weight='bold', color='white')

            # Title: a friendly, narrative reminder
            ax.set_title("Electric field: arrows show direction & strength; lines follow the arrows")

            plt.show()

    # tie controls
    for w in (preset, extent, gridN, density, show_vec, show_lines, show_mag, normalize, clip_r):
        w.observe(recompute_and_draw, names='value')

    # initial draw
    recompute_and_draw()

    # layout
    row1 = HBox([preset, extent, gridN], layout=Layout(justify_content='space-between'))
    row2 = HBox([density, clip_r], layout=Layout(justify_content='flex-start'))
    row3 = HBox([show_vec, normalize, show_lines, show_mag], layout=Layout(justify_content='flex-start'))
    header = Label("🧭 Visualizing E-fields: vectors and ‘lines that follow the arrows’")
    return VBox([header, row1, row2, row3, out])

ui = build_electric_field_demo()
display(ui)  # display the widget box
```
